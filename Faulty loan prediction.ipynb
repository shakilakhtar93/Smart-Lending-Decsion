```python
from imblearn.over_sampling import SMOTE
from sklearn.model_selection import train_test_split
X = data.drop('Risk_Flag', axis=1)
y = data['Risk_Flag']
smote = SMOTE(sampling_strategy='auto', random_state=42)
X_resampled, y_resampled = smote.fit_resample(X, y)
```

```python
y_resampled
```

```python
from sklearn.datasets import make_classification
```

```python
from collections import Counter
```

```python
from sklearn.model_selection import train_test_split
X_train, X_validation, Y_train, Y_validation = train_test_split(X_resampled, y_resampled,
                                                               test_size=0.2, random_state=4)
```

```python
X_train
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
```

```python
lr.fit(X_train, Y_train)
```

```python
Y_predict = lr.predict(X_validation)
```

```python
lr.score(X_validation, Y_validation)
```

```python
from sklearn.metrics import accuracy_score, precision_score, f1_score, recall_score
```

```python
accuracy = accuracy_score(Y_validation, Y_predict)
precision = precision_score(Y_validation, Y_predict)
recall = recall_score(Y_validation, Y_predict)
f1 = f1_score(Y_validation, Y_predict)
print("ACCURACY=", accuracy)
print("PRECESION=", accuracy)
print("RECALL=", accuracy)
print("F1_SCORE=", accuracy)
```
